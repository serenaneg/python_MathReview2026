## 1. Import packages
A **package** is a toolbox of ready-made code. Import at the top.

```python
import numpy as np              # numbers & arrays
import matplotlib.pyplot as plt # plotting
print("packages loaded")
```

## 2. Variables and types
A variable stores a value. `type()` tells you what kind it is.

```python
name = "Woods Hole"   # text (string)
depth = 45            # whole number (int)
temp = 12.5           # decimal (float)

print(name, depth, temp)
print(type(depth))
```

## 3. Math
Python is a calculator. `**` is power, `%` is remainder.

```python
print(2 + 3)
print(10 / 4)
print(2 ** 3)
print(17 % 5)
```

## 4. Lists
A list holds many values in order. Counting starts at **0**.

```python
depths = [10, 25, 40, 55]
print(depths[0])    # first -> 10
print(depths[-1])   # last  -> 55
print(len(depths))  # how many
depths.append(70)   # add to the end
print(depths)
```

## 5. A loop
Do something once for each item.

```python
for d in depths:
    print("depth:", d, "m")
```

## 6. A condition
Do different things depending on a test.

```python
d = 55
if d > 50:
    print("deep")
else:
    print("shallow")
```

## 7. A function
A reusable recipe: give it input, get output back.

```python
def to_fahrenheit(celsius):
    return celsius * 9/5 + 32

print(to_fahrenheit(12.5))
```

## 8. NumPy arrays
Like a list, but built for math on many numbers at once.

```python
a = np.array([10, 25, 40, 55])
print(a.mean())   # average
print(a.max())    # largest
print(a * 2)      # every value doubled
```

## 9. A quick plot

```python
x = np.array([0, 1, 2, 3, 4])
y = x ** 2

plt.plot(x, y, marker="o")
plt.xlabel("x")
plt.ylabel("y = x squared")
plt.title("My first plot")
plt.show()
```
